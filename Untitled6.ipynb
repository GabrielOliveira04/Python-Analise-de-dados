```python
import pandas as pd

# Carregar o arquivo Excel para o DataFrame
tabela = pd.read_excel(r"uber_deliveryServiceDataset.xlsx")

# Filtrar apenas os registros da décima terceira hora
hora_13 = tabela[tabela['Hour'] == 13]

# Contar o número total de produtos na décima terceira hora
total_produtos_hora_13 = hora_13.shape[0]

# Calcular a porcentagem de produtos entregues na décima terceira hora
porcentagem = (total_produtos_hora_13 / tabela.shape[0]) * 100

# Exibir a porcentagem de produtos entregues na décima terceira hora
print(f"A porcentagem de produtos entregues na décima terceira hora é de: {porcentagem:.2f}%")

display(tabela)
```
