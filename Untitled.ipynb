```python
#Agora vamos analisar o DataFrame, vou deixar mais organizado

import pandas as pd

vendas_df = pd.read_csv(r'Contoso - Vendas - 2017.csv', sep=';')
produtos_df = pd.read_csv(r'Contoso - Cadastro Produtos.csv', sep=';')
lojas_df = pd.read_csv(r'Contoso - Lojas.csv',sep=';')
clientes_df = pd.read_csv(r'Contoso - Clientes.csv ', sep=';')


#Segunda Maneira, Pois só quero algumas colunas, só duas colunas, que vamos redefirnir
clientes_df = clientes_df[['ID Cliente', 'E-mail']]
produtos_df = produtos_df[['ID Produto', 'Nome do Produto']]
lojas_df = lojas_df[['ID Loja', 'Nome da Loja']]
#Juntando os dataframes em um só para ficar mais organizado usamos a função merge

vendas_df = vendas_df.merge(produtos_df, on='ID Produto')
vendas_df = vendas_df.merge(lojas_df, on='ID Loja')
vendas_df = vendas_df.merge(clientes_df, on = 'ID Cliente').rename(columns={'E-mail' : 'E-mail do Cliente'})

#Exibindo o novo dataFrame
display(vendas_df)
```

```python
#Qual cliente comprou mais vezes?
#Para isso vou utilizar um metodo e por meio do grafico:
#O método é o .value_counts( para contar quantas vezes cada valor o DF aparece
#E o grafico vou usar o metodo.plot()
#Nessa parte que boto nome da coluna é que eu escolho qual coluna eu quero pegar para fazer essa contagem.
frequencia_clientes = vendas_df['E-mail do Cliente'].value_counts()
display(frequencia_clientes)

#frequencia_clientes.plot()
#Eu chamando ele desse jeito ele está mostrando Length: 38907, porém quero pegar apenas os primeiros 5 itens
#frequencia_clientes[:5].plot()
#porém ele está me dando muito pequeno esse meu gráfico e para aumentar vou usar uma função chamada:
#figsize que nela vou passar uma tupla com largura e altura.

#frequencia_clientes[:5].plot(figsize=(15,5))

#E ai quero que meu eixo comece de forma mais organizado, ai uso a função YTICKS e nele passo 3 parâmetros, como ele começa, até quando que ele vai e de quantos em quantos.

frequencia_clientes[:5].plot(figsize=(15,5), yticks=range(0,100,5))
```

```python
#Qual a loja que mais vendeu em quantidade?
#Usaremos o .groupby poara   agrupar o nosso dataframe, de acordo com o que queremos(somando as quantidades de vendas).
vendas_lojas = vendas_df.groupby('Nome da Loja').sum()
#porém se eu quiser pegar apenas uma coluna como por exemplo quantidade vendida.
vendas_lojas = vendas_lojas[['Quantidade Vendida']]
display(vendas_lojas)
```

```python
#Agora eu quero saber qual foi a loja que mais vendeu e para isso tenho que ordenar e ai tenho dos métodos.
#Método .sort_values: Aonde ordena o df em ordem decrescente usando o metodo ascending = False de Quantidade Vendida.
#E tem os métodos .max() e .idxmax() Que pegar o maior valor diretamente.

#Ordenando o DF
vendas_lojas = vendas_lojas.sort_values('Quantidade Vendida', ascending = False)
display(vendas_lojas)

#Podemos plotar
vendas_lojas[:5].plot(figsize=(15,5), kind='bar')
```

```python
#Pegando o maior valor e o índice
maior_valor = vendas_lojas['Quantidade Vendida'].max()
melhor_loja = vendas_lojas['Quantidade Vendida'].idxmax()
print(maior_valor, melhor_loja)
```

```python
#Qual produto que menos vendeu?
#E temos dois métodos .min() e o idxmin().


menor_valor = vendas_lojas['Quantidade Vendida'].min()
pior_loja = vendas_lojas['Quantidade Vendida'].idxmin()
print(menor_valor,pior_loja)

#Ou posso usar um metodo mais facil já que está ordenado nossa var vendas_lojas:
#Que é apenas pegar o ultimo item vendas_lojas[-1:] que vai da o mesmo resultado
```
